## Exercise 1
Clean the data in the DataFrame **Energy**. The detailed steps are as follows.  
  
#### Steps:   
1. For all the countries which have missing data (e.g., data with '...'), represent them as np.NaN values.
2. Rename the following list of countries:
        "Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"


```python
energy = energy.replace('...',np.nan)

energy['Country Name'] = energy['Country Name'].replace({
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
})


```

####  Execute the following statements
1. Because of '...' in the original data, the columns 'Energy Supply' and 'Energy Supply per capita' are recognized as strings. Use **astype** to convert them to be floating point number.
2. "Energy Supply" is in petajoules but "Energy Supply per capita" is in gigajoules. Convert the "Energy Supply" to gigajoules (there are 1,000,000 gigajoules in a petajoule).

```python
energy['Energy Supply'] = energy['Energy Supply'].astype(float)
energy['Energy Supply per capita'] = energy['Energy Supply per capita'].astype(float)
energy["Energy Supply"] = energy["Energy Supply"]*1000000
```
